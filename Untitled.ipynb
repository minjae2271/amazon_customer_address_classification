```python
df = pd.read_csv('Account status_2024.02.csv')
```

```python
df = df.rename(columns={'account_billing_address_city_code': 'city', 'account_billing_address_state_code': 'state'})
df = df[['account_id', 'city', 'state']]

df = df.applymap(lambda x: str(x).replace(' ', '').replace('-', ''))

print(df.head())
print('count: ', df.count())
```

```python
no_seoul_gyeonggi_df = df[(df['city'].str.upper() == 'SEOUL') | (df['state'].str.upper() == 'SEOUL') | (df['city'].str.upper() == 'GYEONGGI-DO') | (df['state'].str.upper() == 'GYEONGGI-DO')]
print(no_seoul_gyeonggi_df.count())
print(no_seoul_gyeonggi_df.head())
```

```python
non_urban = df.drop(no_seoul_gyeonggi_df.index)

print("Urban DataFrame:")
print(no_seoul_gyeonggi_df.count())

print("\nNon-Urban DataFrame:")
print(non_urban.count())
```

```python
#city columns의 distinct
value_counts_non_urban_city = non_urban['city'].value_counts()
print("\nNon-Urban city Value Counts:")
print(value_counts_non_urban_city.count())
print(value_counts_non_urban_city.head((30)))

#state columns의 distinct
value_counts_non_urban_state = non_urban['state'].value_counts()
print("\nNon-Urban State Value Counts:")
print(value_counts_non_urban_state.count())
print(value_counts_non_urban_state.head(10))
```

```python
# Query to filter rows based on conditions
filtered_df = df[~(df['city'].fillna('').str.contains('|'.join(seoul_districts + gyeonggi_cities + province)) | df['state'].fillna('').str.contains('|'.join(seoul_districts + gyeonggi_cities + province)))]

# Drop rows with all null values
filtered_df.replace('#N/A#', np.nan, inplace=True)
res_df = filtered_df.dropna(subset=['city', 'state'], how='all')

# Display the resulting DataFrame
print(res_df.count())
print(res_df.head())
```
