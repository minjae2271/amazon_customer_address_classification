```python
df = pd.read_csv('Account status_2024.02.csv')
```

```python
df = df.rename(columns={'account_billing_address_city_code': 'city', 'account_billing_address_state_code': 'state'})
df = df[['account_id', 'city', 'state']]

df = df.applymap(lambda x: str(x).replace(' ', '').replace('-', ''))

print(df.head())
print('count: ', df.count())
```

```python
no_seoul_gyeonggi_df = df[(df['city'].str.upper() == 'SEOUL') | (df['state'].str.upper() == 'SEOUL') | (df['city'].str.upper() == 'GYEONGGI-DO') | (df['state'].str.upper() == 'GYEONGGI-DO')]
print(no_seoul_gyeonggi_df.count())
print(no_seoul_gyeonggi_df.head())
```

```python
non_urban = df.drop(no_seoul_gyeonggi_df.index)

print("Urban DataFrame:")
print(no_seoul_gyeonggi_df.count())

print("\nNon-Urban DataFrame:")
print(non_urban.count())
```

```python
#city columns의 distinct
value_counts_non_urban_city = non_urban['city'].value_counts()
print("\nNon-Urban city Value Counts:")
print(value_counts_non_urban_city.count())
print(value_counts_non_urban_city.head((30)))

#state columns의 distinct
value_counts_non_urban_state = non_urban['state'].value_counts()
print("\nNon-Urban State Value Counts:")
print(value_counts_non_urban_state.count())
print(value_counts_non_urban_state.head(10))
```

```python
# Create the 'region' column
rural_df['region'] = None

# Create a list of lists
#regions_lists = [gangwon, chungnam, chungbuk, jeju, jeonnam, jeonbuk, gyeongnam, gyeongbuk]

def assign_region(row):
    regions = {
        'gangwon': gangwon,
        'chungnam': chungnam,
        'chungbuk': chungbuk,
        'jeju': jeju,
        'jeonnam': jeonnam,
        'jeonbuk': jeonbuk,
        'gyeongnam': gyeongnam,
        'gyeongbuk': gyeongbuk,
    }
    city = str(row['city']) if pd.notnull(row['city']) else None
    state = str(row['state']) if pd.notnull(row['state']) else None

    for region_name, region_list in regions.items():
        # Check if city is not None and matches any region
        if city and any(s in city for s in region_list):
            return region_name
        # Check if state matches any region
        elif state and any(s in state for s in region_list):
            return region_name
    return np.nan

# Apply the function
rural_df['region'] = rural_df.apply(assign_region, axis=1)

# Display the updated DataFrame
print(rural_df.count())
print(rural_df.head())
rural_df.to_csv('rural_csv.csv', index=False)
rural_df.to_excel('rural_excel.xlsx', index=False)
```
